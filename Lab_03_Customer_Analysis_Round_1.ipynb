# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
set i=0 to iterate over the entries.

iterate over the values in customer_ltv column.

change the value to a string and replace/remove the '%' at the end.

convert string into float, multiple by 100 and round to two decimals.

```python
i=0
for value in df4.customer_ltv:
    df4.customer_ltv[i] = round(float(str(value).replace('%',''))*100,2)
    i+=1
display(df4)
```
